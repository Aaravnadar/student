## 6) Analytics review

Reviewing our project analytics was eye-opening. The most interesting discovery was seeing the drop-off rate on certain pages of our blog and quiz sections. We found that users would often leave after the initial introductory page. This showed us that even if the code works, the content needs to be compelling and structured to immediately hook the user. We also saw that the time spent on our gamification page was significantly higher than on informational pages, confirming that interactive elements are key to engagement. This review confirmed the importance of using data to guide future design decisions.
