## My Growth: Then vs. Now (2025)

### 1) Where I started
At the start of the year I knew how to write basic code, but working inside a full repository with notebooks, build scripts, and a Makefile felt messy. I struggled with:
- Getting dependencies installed and using virtual environments.
- Debugging Python npm/JS module errors.
- Understanding how notebook conversions (nbconvert) and the Makefile interacted with the site build.
- Committing clean, meaningful git changes.

### 2) What I can do now
Over the year I learned concrete, practical skills:
- Environment and dependency management: I create virtual environments, install packages locally, and troubleshoot ModuleNotFoundError issues (for example installing `nbconvert` and `progress`).
- Build troubleshooting: I can read `make` errors, identify missing directories or packages, and either create missing folders or update the Makefile steps.
- JavaScript + Notebook work: I can write interactive `%%html` cells, add DOM elements, and wire up event listeners for small assignments (Popcorn hacks, homework functions).
- Git basics: I stage changes, write commit messages, and push to the remote repo.

### 3) Concrete wins
- Fixed a notebook conversion error by installing `nbconvert` in the virtual environment and creating missing `_notebooks/CSP` directories when needed.
- Implemented and debugged multiple JS homework tasks: math functions, conditional hacks, array utilities.
- Built a small CookieClicker class with upgrades — learned how to structure classes and test usage in the notebook.

### 4) What’s next
- Add a `requirements.txt` or pinned environment file for reproducibility.
- Refactor the Makefile to skip non-existent folders or make the script robust to missing notebooks.
- Write at least one polished notebook blog post showing a project from idea → code → notes (and publish it).
- Start writing small tests or assertions for notebook outputs so grading/review is faster.

### 5) Personal note
The biggest difference is confidence: I know where to look when things break, how to search/diagnose, and how to take small steps until the build is green. That’s a change that makes learning new features much faster.
